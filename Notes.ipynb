# Determinant and Inverse of a Matrix

## Determinant of a Matrix

The determinant of a matrix is a scalar value that is a function of the entries of a square matrix. It provides important properties of the matrix and can be used to determine whether a matrix is invertible.

### Intuition
- **Geometric Interpretation**: The determinant of a 2x2 matrix can be interpreted as the area of the parallelogram formed by its column vectors. For higher dimensions, it represents the volume of the parallelepiped.
- **Properties**:
  - If the determinant is zero, the matrix is singular (non-invertible).
  - The sign of the determinant can indicate the orientation of the transformation described by the matrix.

### Calculation
For a 2x2 matrix:
$ A = \begin{pmatrix} a & b \\ c & d \end{pmatrix} $
The determinant is calculated as:
$ \text{det}(A) = ad - bc $

For larger matrices, the determinant is calculated using more complex methods such as cofactor expansion.

## Inverse of a Matrix

The inverse of a matrix $ A $ is another matrix $ A^{-1} $ such that:
$ A A^{-1} = A^{-1} A = I $
where $ I $ is the identity matrix.

### Intuition
- **Geometric Interpretation**: The inverse of a matrix represents a transformation that "undoes" the transformation represented by the original matrix.
- **Properties**:
  - A matrix is invertible if and only if its determinant is non-zero.
  - The inverse of a product of matrices is the product of their inverses in reverse order.

### Calculation
For a 2x2 matrix:
$ A = \begin{pmatrix} a & b \\ c & d \end{pmatrix} $
The inverse is calculated as:
$ A^{-1} = \frac{1}{\text{det}(A)} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix} $
provided that $ \text{det}(A) \neq 0 $.

## Example in Python


```python
import numpy as np

# Define a 2x2 matrix
A = np.array([[4, 7], [2, 6]])

# Calculate the determinant
det_A = np.linalg.det(A)
print(f"Determinant of A: {det_A}")

# Calculate the inverse
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print(f"Inverse of A:\n{inv_A}")
else:
    print("Matrix A is singular and does not have an inverse.")
```

## Use in Gaussian Distribution

In the context of the multivariate Gaussian distribution, the determinant and inverse of the covariance matrix $ \Sigma $ play crucial roles.

### Determinant
- **Volume Scaling**: The determinant of $ \Sigma $ scales the volume of the distribution. A larger determinant indicates a more spread out distribution, while a smaller determinant indicates a more concentrated distribution.

### Inverse
- **Mahalanobis Distance**: The inverse of $ \Sigma $ is used to compute the Mahalanobis distance, which measures the distance between a point and the mean of the distribution, taking into account the correlations between variables.

### Probability Density Function (PDF)
The PDF of a multivariate Gaussian distribution is given by:
$ f(\mathbf{x}) = \frac{1}{(2\pi)^{k/2} |\Sigma|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x} - \mu)^T \Sigma^{-1} (\mathbf{x} - \mu)\right) $

where:
- $ \mathbf{x} $ is a $ k $-dimensional random vector.
- $ \mu $ is the mean vector.
- $ \Sigma $ is the covariance matrix.
- $ |\Sigma| $ is the determinant of the covariance matrix.
- $ \Sigma^{-1} $ is the inverse of the covariance matrix.

The determinant $ |\Sigma| $ normalizes the distribution, while the inverse $ \Sigma^{-1} $ adjusts for the correlations between variables.

# Gaussian Distribution

## Introduction
The Gaussian distribution, also known as the normal distribution, is a continuous probability distribution that is symmetrical around its mean. It is characterized by its bell-shaped curve.

## Properties
1. **Mean** ($\mu$) : The central value around which the distribution is symmetrical.
2. **Standard Deviation** ($\sigma$): Measures the spread of the distribution. A smaller standard deviation indicates that the data points are closer to the mean, while a larger standard deviation indicates that the data points are spread out over a wider range of values.
3. **Variance** ($\sigma^2$): The square of the standard deviation.

## Probability Density Function (PDF)
The PDF of a Gaussian distribution is given by:

$ f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} $

## Example

```python
#Let's generate a Gaussian distribution using Python and visualize it.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate data
x = np.linspace(-5, 5, 1000)
y = norm.pdf(x, mu, sigma)

# Plot
plt.plot(x, y)
plt.title('Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```

# Multivariate Gaussian Distribution

## Introduction
The multivariate Gaussian distribution is a generalization of the one-dimensional (univariate) Gaussian distribution to higher dimensions. It describes the joint distribution of a set of random variables that may be correlated with each other.

## Intuition
- **Univariate vs. Multivariate**: While the univariate Gaussian distribution describes a single random variable, the multivariate Gaussian distribution describes multiple random variables simultaneously.
- **Correlation**: In the multivariate case, the variables can be correlated, meaning that the value of one variable can give information about the value of another.

## Properties
1. **Mean Vector ($\mu$)**: A vector of means for each variable.
2. **Covariance Matrix ($\Sigma$)**: A matrix that describes the variance of each variable along the diagonal and the covariance between variables off the diagonal.

## Probability Density Function (PDF)
The PDF of a multivariate Gaussian distribution is given by:

$ f(\mathbf{x}) = \frac{1}{(2\pi)^{k/2} |\Sigma|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x} - \mu)^T \Sigma^{-1} (\mathbf{x} - \mu)\right) $
where:
- $\mathbf{x}$ is a $k$-dimensional random vector, representing the variables.
    - k means the number of variables.
    - $k=2$ represents a bivariate distribution.
- $\mu$ is the mean vector.
- $\Sigma$ is the covariance matrix.
- $|\Sigma|$ is the determinant of the covariance matrix.
- $\Sigma^{-1}$ is the inverse of the covariance matrix.

## Example
Let's generate a multivariate Gaussian distribution using Python and visualize it.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parameters
mu = np.array([0, 0])  # Mean vector
sigma = np.array([[1, 0.5], [0.5, 1]])  # Covariance matrix

# Generate grid
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

# Multivariate normal distribution
rv = multivariate_normal(mu, sigma)
z = rv.pdf(pos)

# Plot
plt.contourf(x, y, z, cmap='viridis')
plt.title('Multivariate Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

# Parametric Classifiers

## Introduction
Parametric classifiers are a type of classification algorithm that make assumptions about the underlying distribution of the data. These classifiers use a fixed number of parameters to describe the data, which simplifies the learning process and often leads to faster and more efficient algorithms.

## Key Characteristics
1. **Assumptions**: They assume a specific form for the distribution of the data (e.g., Gaussian distribution).
2. **Fixed Parameters**: The number of parameters is fixed and does not grow with the size of the dataset.
3. **Efficiency**: They are generally computationally efficient due to the fixed number of parameters.
4. **Examples**: Common examples include Linear Discriminant Analysis (LDA), Quadratic Discriminant Analysis (QDA), and Gaussian Naive Bayes.

# Gaussian-based Classifiers

## Introduction
Gaussian-based classifiers are a subset of parametric classifiers that assume the data follows a Gaussian (normal) distribution. These classifiers are particularly useful when the data is continuous and can be well-approximated by a Gaussian distribution.

## Gaussian Naive Bayes
Gaussian Naive Bayes is a probabilistic classifier based on Bayes' theorem with the assumption that the features are normally distributed.

### Assumptions
1. **Feature Independence**: It assumes that the features are independent given the class label.
2. **Gaussian Distribution**: It assumes that the features follow a Gaussian distribution.

### Probability Density Function (PDF)
The PDF of a Gaussian distribution for a feature $ x_i $ is given by:
$ f(x_i) = \frac{1}{\sqrt{2\pi\sigma_i^2}} \exp\left(-\frac{(x_i - \mu_i)^2}{2\sigma_i^2}\right) $
where:
- $ \mu_i $ is the mean of the feature.
- $ \sigma_i^2 $ is the variance of the feature.

### Classification Rule
The classifier assigns a class label $ y $ to a new data point $ \mathbf{x} $ by maximizing the posterior probability:
$ y = \arg\max_y P(y) \prod_{i=1}^n f(x_i | y) $
where:
- $ P(y) $ is the prior probability of class $ y $.
- $ f(x_i | y) $ is the likelihood of feature $ x_i $ given class $ y $.

## Example in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X = np.random.randn(100, 2)
y = np.array([0 if x[0] + x[1] < 0 else 1 for x in X])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Initialize the Gaussian Naive Bayes classifier
gnb = GaussianNB()

# Fit the model
gnb.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = gnb.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = gnb.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.title('Gaussian Naive Bayes Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Quadratic Discriminant Analysis (QDA)

## Intuition

Quadratic Discriminant Analysis (QDA) is a classification algorithm that assumes that the data from each class is drawn from a Gaussian distribution. Unlike Linear Discriminant Analysis (LDA), which assumes that each class shares the same covariance matrix, QDA allows each class to have its own covariance matrix. This flexibility allows QDA to model more complex decision boundaries.

### Discriminant Function

This function combines the likelihood of the data point belonging to the class, the prior probability of the class, and the covariance structure of the class.

The discriminant function for QDA is given by:
$ \delta_k(\mathbf{x}) = -\frac{1}{2} \log |\Sigma_k| - \frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma_k^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $
where:

- $ \delta_k(\mathbf{x}) $ is the discriminant function for class $ k $.
- $ \mathbf{x} $ is the input data.
- $ \mu_k $ is the mean vector of class $ k $.
- $ \Sigma_k $ is the covariance matrix of class $ k $.
- P(y = k) is the **prior probability** of class $ k $.

#### General Form
$ f(\mathbf{x}) = \mathbf{x}^T W \mathbf{x} + \mathbf{w}^T \mathbf{x} + w_0 $
- Quadratic term: $ \mathbf{x}^T W \mathbf{x} $
    - This term captures the curvature of the decision boundary.
- Linear term: $ \mathbf{w}^T \mathbf{x} $
    - This term captures the influence of the input features.
    - This term captures the linear relationship between the features and the class.
- Constant term: $ w_0 $
    - This term adjusts the decision boundary based on the class priors and the covariance structure.
 

### Key Characteristics
1. **Class-Specific Covariance Matrices**: Each class has its own covariance matrix, allowing for more flexible decision boundaries.
2. **Quadratic Decision Boundaries**: The decision boundaries between classes are quadratic (parabolic) rather than linear.
3. **Gaussian Assumption**: Assumes that the data for each class follows a Gaussian distribution.

## Best Fit Data

QDA is best suited for:
- **Data with Non-Linear Boundaries**: When the decision boundaries between classes are not linear.
- **Sufficient Data**: Requires a relatively large amount of data to accurately estimate the covariance matrices for each class.
- **Continuous Features**: Works well with continuous data that can be approximated by a Gaussian distribution.

## Relevance of Probabilities in QDA

In the context of classification, particularly with Quadratic Discriminant Analysis (QDA), the probabilities $ P(x) $, $ P(y) $, $ P(x|y) $, and $ P(y|x) $ play specific roles:

1. **$ P(x) $ - Marginal Probability of $ x $**:
   - **Relevance**: This is the overall probability of observing the feature vector $ x $ in the dataset.
   - **Known or Estimated**: It is generally not directly estimated in QDA. Instead, it is implicitly considered when normalizing probabilities.
   - **Calculation**: Not directly calculated in QDA.

2. **$ P(y) $ - Prior Probability of Class $ y $**:
   - **Relevance**: This is the probability of a data point belonging to class $ y $ before observing the feature vector $ x $.
   - **Known or Estimated**: Estimated from the training data as the proportion of data points in each class.
   - **Calculation**: Estimated during the training phase.

3. **$ P(x|y) $ - Likelihood of $ x $ Given Class $ y $**:
   - **Relevance**: This is the probability of observing the feature vector $ x $ given that it belongs to class $ y $. In QDA, this is modeled using a Gaussian distribution with class-specific mean and covariance.
   - **Known or Estimated**: Estimated from the training data by fitting a Gaussian distribution to the data points in each class.
   - **Calculation**: Estimated during the training phase.

4. **$ P(y|x) $ - Posterior Probability of Class $ y $ Given $ x $**:
   - **Relevance**: This is the probability of a data point belonging to class $ y $ after observing the feature vector $ x $. This is what we aim to calculate when classifying a new data point.
   - **Known or Estimated**: Calculated using Bayes' theorem during the classification phase.
   - **Calculation**: Calculated when a new $ x $ needs to be classified.

### Summary of Estimation and Calculation

- **Estimated from Training Data**:
  - $ P(y) $
  - $ P(x|y) $ (through the mean vector $ \mu_k $ and covariance matrix $ \Sigma_k $)

- **Calculated During Classification**:
  - $ P(y|x) $

### Bayes' Theorem in Classification

To classify a new data point $ x $, we use Bayes' theorem:
$ P(y|x) = \frac{P(x|y) \cdot P(y)}{P(x)} $

In practice, we compare the discriminant functions for each class:
$ \delta_k(\mathbf{x}) = -\frac{1}{2} \log |\Sigma_k| - \frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma_k^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $

### Why Use the Discriminant Function?
Avoiding $ P(x) $: Directly calculating $ P(x) $ can be complex and unnecessary for classification. Since $ P(x) $ is the same for all classes, it cancels out when comparing classes.
Computational Efficiency: The discriminant function combines the necessary terms in a way that is computationally efficient and avoids the need to explicitly calculate $P(x)$.



The class with the highest discriminant function value is chosen as the predicted class.

Quadratic Discriminant Analysis (QDA) is a powerful classification algorithm that models class-specific covariance matrices, allowing for quadratic decision boundaries. It is best suited for data with non-linear boundaries and requires a sufficient amount of data to accurately estimate the parameters. Understanding the underlying Gaussian assumptions and the discriminant function is crucial for effectively applying QDA to classification problems.

## How It Works

1. **Estimate Parameters**: For each class, estimate the mean vector $ \mu_k $ and the covariance matrix $ \Sigma_k $.
2. **Compute Discriminant Function**: For a new data point $ \mathbf{x} $, compute the discriminant function for each class.
3. **Assign Class**: Assign the class label that maximizes the discriminant function.


### Quadratic Decision Boundaries

The term "quadratic" in QDA refers to the fact that the decision boundaries between classes are quadratic functions of the input features. This is due to the class-specific covariance matrices, which result in a quadratic form in the discriminant function.

## Formulas

### Mean Vector and Covariance Matrix Estimation
For each class $ k $:
- Mean vector $ \mu_k $:
$ \mu_k = \frac{1}{N_k} \sum_{i \in C_k} \mathbf{x}_i $
where $ N_k $ is the number of samples in class $ k $ and $ \mathbf{x}_i $ are the samples in class $ k $.

- Covariance matrix $ \Sigma_k $:
$ \Sigma_k = \frac{1}{N_k - 1} \sum_{i \in C_k} (\mathbf{x}_i - \mu_k)(\mathbf{x}_i - \mu_k)^T $

### Discriminant Function
The discriminant function for class $ k $ is given by:
$ \delta_k(\mathbf{x}) = -\frac{1}{2} \log |\Sigma_k| - \frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma_k^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $
where:
- $ |\Sigma_k| $ is the determinant of the covariance matrix for class $ k $.
- $ \Sigma_k^{-1} $ is the inverse of the covariance matrix for class $ k $.
- $ P(y = k) $ is the prior probability of class $ k $.

### Classification Rule
Assign the class label $ y $ to a new data point $ \mathbf{x} $ by maximizing the discriminant function:
$ y = \arg\max_k \delta_k(\mathbf{x}) $

## Example in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X = np.random.randn(100, 2)
y = np.array([0 if x[0] + x[1] < 0 else 1 for x in X])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Initialize the Gaussian Naive Bayes classifier
gnb = GaussianNB()

# Fit the model
gnb.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = gnb.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Plotting decision boundary in 2D
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = gnb.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.title('2D Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Plotting decision boundary in 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], y, c=y, marker='o')

# Create a meshgrid for 3D plot
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
zz = gnb.predict(np.c_[xx.ravel(), yy.ravel()])
zz = zz.reshape(xx.shape)

# Plot the surface
ax.plot_surface(xx, yy, zz, alpha=0.3, rstride=100, cstride=100)

ax.set_title('3D Decision Boundary')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Class')
plt.show()
```

# Linear Discriminant Analysis (LDA) vs. Quadratic Discriminant Analysis (QDA)

## Introduction to Linear Discriminant Analysis (LDA)
Linear Discriminant Analysis (LDA) is a classification algorithm that assumes that the data from each class is drawn from a Gaussian distribution with the same covariance matrix. This results in linear decision boundaries between classes.

### Key Characteristics of LDA
1. **Shared Covariance Matrix**: All classes share the same covariance matrix.
2. **Linear Decision Boundaries**: The decision boundaries between classes are linear.
3. **Gaussian Assumption**: Assumes that the data for each class follows a Gaussian distribution.

### Discriminant Function for LDA
The discriminant function for LDA is given by:
    $ \delta_k(\mathbf{x}) = \mathbf{x}^T \Sigma^{-1} \mu_k - \frac{1}{2} \mu_k^T \Sigma^{-1} \mu_k + \log P(y = k) $
where:
- $ \delta_k(\mathbf{x}) $ is the discriminant function for class $ k $.
- $ \mathbf{x} $ is the input data.
- $ \mu_k $ is the mean vector of class $ k $.
- $ \Sigma $ is the shared covariance matrix.
- $ P(y = k) $ is the prior probability of class $ k $.

## Introduction to Quadratic Discriminant Analysis (QDA)
Quadratic Discriminant Analysis (QDA) is a classification algorithm that assumes that the data from each class is drawn from a Gaussian distribution. Unlike LDA, QDA allows each class to have its own covariance matrix. This flexibility allows QDA to model more complex decision boundaries.

### Key Characteristics of QDA
1. **Class-Specific Covariance Matrices**: Each class has its own covariance matrix, allowing for more flexible decision boundaries.
2. **Quadratic Decision Boundaries**: The decision boundaries between classes are quadratic (parabolic) rather than linear.
3. **Gaussian Assumption**: Assumes that the data for each class follows a Gaussian distribution.

## Intuition Behind Averaging Covariance Matrices
When there is insufficient data, the covariance matrix for a class may become singular (non-invertible). This can happen if one of the variances is zero, indicating that there is no variability in one of the features for that class. To address this issue, we can average the covariance matrices across all classes, effectively pooling the data to create a more stable estimate of the covariance.

### Is This Still QDA?
No, this approach is more akin to Linear Discriminant Analysis (LDA). In LDA, we assume that all classes share the same covariance matrix, which is estimated by averaging the covariance matrices of all classes. This results in linear decision boundaries rather than quadratic ones.

### Using the Averaged Covariance Matrix
When using the averaged covariance matrix, the same covariance matrix is used for all class discriminant functions. This means that the only class-specific parameters are the means.

### Formulas

#### Averaged Covariance Matrix
The averaged covariance matrix $ \Sigma $ is calculated as:
$ \Sigma = \frac{1}{K} \sum_{k=1}^{K} \Sigma_k $
where $ K $ is the number of classes and $ \Sigma_k $ is the covariance matrix for class $ k $.

#### Discriminant Function
The discriminant function for class $ k $ in this case is:
    $ \delta_k(\mathbf{x}) = -\frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $
where:
- $ \mu_k $ is the mean vector for class $ k $.
- $ \Sigma $ is the averaged covariance matrix.
- $ P(y = k) $ is the prior probability of class $ k $.

### Summary
- **Averaging Covariance Matrices**: Used when individual class covariance matrices are singular or unstable.
- **LDA vs. QDA**: Averaging leads to LDA, where all classes share the same covariance matrix, resulting in linear decision boundaries.
- **Discriminant Function**: Uses the averaged covariance matrix, making the means the primary class-specific parameters.

# Two-Class Example: Linear Discriminant Analysis (LDA)

## Introduction
In this example, we will explore Linear Discriminant Analysis (LDA) for a two-class classification problem. LDA is a classification algorithm that assumes that the data from each class is drawn from a Gaussian distribution with the same covariance matrix. This results in linear decision boundaries between classes.

## Key Concepts

### Mean Vectors
For each class $ k $, the mean vector $ \mu_k $ is calculated as:
$ \mu_k = \frac{1}{N_k} \sum_{i \in C_k} \mathbf{x}_i $
where $ N_k $ is the number of samples in class $ k $ and $ \mathbf{x}_i $ are the samples in class $ k $.

### Covariance Matrices
For each class $ k $, the covariance matrix $ \Sigma_k $ is calculated as:
$ \Sigma_k = \frac{1}{N_k - 1} \sum_{i \in C_k} (\mathbf{x}_i - \mu_k)(\mathbf{x}_i - \mu_k)^T $

### Averaged Covariance Matrix
The averaged covariance matrix $ \Sigma $ is calculated as:
$ \Sigma = \frac{1}{2} (\Sigma_1 + \Sigma_2) $
where $ \Sigma_1 $ and $ \Sigma_2 $ are the covariance matrices for class 1 and class 2, respectively.

### Discriminant Function
The discriminant function for class $ k $ is given by:
$ \delta_k(\mathbf{x}) = -\frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $
where:
- $ \mu_k $ is the mean vector for class $ k $.
- $ \Sigma $ is the averaged covariance matrix.
- $ P(y = k) $ is the prior probability of class $ k $.

### Linear Discriminant Function
The linear discriminant function $ f(\mathbf{x}) $ is defined as:
$ f(\mathbf{x}) = \log P(y=1|\mathbf{x}) - \log P(y=2|\mathbf{x}) $
This can be expressed as:
$ f(\mathbf{x}) = \mathbf{w}^T \mathbf{x} + w_0 $
where:
- $ \mathbf{w} = \Sigma^{-1} (\mu_2 - \mu_1) $
- $ w_0 = -\frac{1}{2} \mu_2^T \Sigma^{-1} \mu_2 + \frac{1}{2} \mu_1^T \Sigma^{-1} \mu_1 $

## Summary
- **Mean Vectors**: Calculated for each class.
- **Covariance Matrices**: Calculated for each class.
- **Averaged Covariance Matrix**: Used to stabilize the covariance estimate.
- **Discriminant Function**: Uses the averaged covariance matrix, resulting in linear decision boundaries.
- **Linear Discriminant Function**: Expressed as a linear combination of the input features.

```python
# Comprehensive Example of LDA and QDA with Visualizations

import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X = np.random.randn(200, 2)
y = np.array([0 if x[0]**2 + x[1]**2 < 1 else 1 for x in X])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Initialize the LDA and QDA classifiers
lda = LinearDiscriminantAnalysis()
qda = QuadraticDiscriminantAnalysis()

# Fit the models
lda.fit(X_train, y_train)
qda.fit(X_train, y_train)

# Predict the labels for the test set
y_pred_lda = lda.predict(X_test)
y_pred_qda = qda.predict(X_test)

# Calculate the accuracy
accuracy_lda = accuracy_score(y_test, y_pred_lda)
accuracy_qda = accuracy_score(y_test, y_pred_qda)

print(f"LDA Accuracy: {accuracy_lda}")
print(f"QDA Accuracy: {accuracy_qda}")

# Plot decision boundaries
def plot_decision_boundary(model, X, y, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plot_decision_boundary(lda, X, y, 'LDA Decision Boundary')

plt.subplot(1, 2, 2)
plot_decision_boundary(qda, X, y, 'QDA Decision Boundary')

plt.tight_layout()
plt.show()
```

## Simplified Covariance Matrix in Gaussian-Based Classification

In some cases, even a full averaged covariance matrix is too complex to estimate accurately, especially when the dataset is small or the number of features is large. A common simplification is to assume that all features have the same variance and are uncorrelated. This results in a diagonal covariance matrix where the off-diagonal elements are zero.

### Simplified Covariance Matrix

The simplified covariance matrix can be estimated by multiplying the standard deviation by the identity matrix. This approach assumes that the variance of each feature is the same and that there are no correlations between features.

### Comparison to Averaged Covariance Matrix

- **Averaged Covariance Matrix**: This matrix is calculated by averaging the covariance matrices of all classes. It captures the variances and covariances of the features, allowing for linear decision boundaries.
- **Simplified Covariance Matrix**: This matrix assumes equal variance for all features and no correlation between them. It simplifies the estimation process and results in a diagonal covariance matrix.

### Derivation of the Discriminant Function

Given the simplified covariance matrix, the discriminant function can be derived as follows:

1. **Covariance Matrix**:
$
\Sigma = \sigma^2 I
$
where $ \sigma^2 $ is the common variance and $ I $ is the identity matrix.

2. **Inverse of Covariance Matrix**:
$
\Sigma^{-1} = \frac{1}{\sigma^2} I
$

3. **Determinant of Covariance Matrix**:
$
 |\Sigma| = (\sigma^2)^d
$
where $ d $ is the number of features.

4. **Discriminant Function**:
$
\delta_k(\mathbf{x}) = -\frac{1}{2} \log |\Sigma| - \frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma^{-1} (\mathbf{x} - \mu_k) + \log P(y = k)
$

Substituting the simplified covariance matrix:
$
\delta_k(\mathbf{x}) = -\frac{1}{2} \log (\sigma^2)^d - \frac{1}{2} (\mathbf{x} - \mu_k)^T \left( \frac{1}{\sigma^2} I \right) (\mathbf{x} - \mu_k) + \log P(y = k)
$

Simplifying further:
$
\delta_k(\mathbf{x}) = -\frac{d}{2} \log \sigma^2 - \frac{1}{2\sigma^2} (\mathbf{x} - \mu_k)^T (\mathbf{x} - \mu_k) + \log P(y = k)
$

Since $ \sigma^2 $ is common for all classes, it can be ignored when comparing discriminant functions:
    $
\delta_k(\mathbf{x}) = - \frac{1}{2\sigma^2} (\mathbf{x} - \mu_k)^T (\mathbf{x} - \mu_k) + \log P(y = k)
$

### Example in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X = np.random.randn(200, 2)
y = np.array([0 if x[0] + x[1] < 0 else 1 for x in X])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Initialize the LDA classifier
lda = LinearDiscriminantAnalysis()

# Fit the model
lda.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = lda.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = lda.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.title('Linear Discriminant Analysis Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Nearest Mean Classifier (NMC)

### Introduction
The Nearest Mean Classifier (NMC) is a simple and intuitive classification algorithm. It assigns a data point to the class whose mean (centroid) is closest to the data point in terms of Euclidean distance. This method is particularly useful when the classes are well-separated and have similar variances.

### Intuition
The intuition behind NMC is straightforward:
1. **Calculate the Mean**: For each class, calculate the mean (centroid) of the data points.
2. **Assign Class**: For a new data point, calculate the distance to each class mean and assign the class with the smallest distance.

### Relevant Formulas
1. **Mean Vector**: For each class $ k $, the mean vector $ \mu_k $ is calculated as:
    $
    \mu_k = \frac{1}{N_k} \sum_{i \in C_k} \mathbf{x}_i
    $
    where $ N_k $ is the number of samples in class $ k $ and $ \mathbf{x}_i $ are the samples in class $ k $.

2. **Distance Calculation**: The Euclidean distance between a data point $ \mathbf{x} $ and the mean vector $ \mu_k $ is given by:
$
    d(\mathbf{x}, \mu_k) = \|\mathbf{x} - \mu_k\| = \sqrt{(\mathbf{x} - \mu_k)^T (\mathbf{x} - \mu_k)}
$

3. **Classification Rule**: Assign the class label $ y $ to a new data point $ \mathbf{x} $ by minimizing the distance:
$
    y = \arg\min_k d(\mathbf{x}, \mu_k)
$

### Comparison to Previous Models
- **Gaussian-Based Classifiers**: These classifiers assume that the data follows a Gaussian distribution and use the covariance matrix to model the data. NMC, on the other hand, only uses the mean of each class and does not consider the covariance.
- **Linear Discriminant Analysis (LDA)**: LDA assumes that all classes share the same covariance matrix and results in linear decision boundaries. NMC results in linear decision boundaries but does not use the covariance matrix.
- **Quadratic Discriminant Analysis (QDA)**: QDA allows each class to have its own covariance matrix, resulting in quadratic decision boundaries. NMC does not use the covariance matrix and results in simpler linear decision boundaries.

### Example in Python


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X_class0 = np.random.randn(100, 2) + np.array([2, 2])
X_class1 = np.random.randn(100, 2) + np.array([-2, -2])
X = np.vstack((X_class0, X_class1))
y = np.array([0] * 100 + [1] * 100)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Calculate the mean vectors for each class
mean_vectors = []
for label in np.unique(y_train):
    mean_vectors.append(np.mean(X_train[y_train == label], axis=0))

# Predict the class for each test point
def predict(X, mean_vectors):
    predictions = []
    for x in X:
        distances = [np.linalg.norm(x - mean) for mean in mean_vectors]
        predictions.append(np.argmin(distances))
    return np.array(predictions)

y_pred = predict(X_test, mean_vectors)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Plot decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = predict(np.c_[xx.ravel(), yy.ravel()], mean_vectors)
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.title('Nearest Mean Classifier Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

## Introduction
Non-parametric classification refers to a set of classification techniques that do not make strong assumptions about the form of the underlying data distribution. Unlike parametric methods, which assume a specific distribution (e.g., Gaussian) and estimate a fixed number of parameters, non-parametric methods are more flexible and can adapt to the structure of the data.

## Key Characteristics
1. **No Assumptions**: Non-parametric methods do not assume a specific form for the data distribution.
2. **Flexibility**: They can model complex relationships in the data.
3. **Data-Driven**: The complexity of the model grows with the amount of data.
4. **Memory-Based**: Often require storing the entire training dataset.

## Comparison to Parametric Classification
- **Assumptions**:
  - **Parametric**: Assumes a specific distribution (e.g., Gaussian) and estimates a fixed number of parameters.
  - **Non-Parametric**: Makes minimal assumptions about the data distribution.
- **Model Complexity**:
  - **Parametric**: Fixed complexity, independent of the dataset size.
  - **Non-Parametric**: Complexity grows with the dataset size.
- **Flexibility**:
  - **Parametric**: Less flexible, may not capture complex patterns.
  - **Non-Parametric**: More flexible, can capture complex patterns.
- **Efficiency**:
  - **Parametric**: Generally more computationally efficient.
  - **Non-Parametric**: Can be computationally intensive, especially with large datasets.

   - **Application**: Image recognition, recommendation systems.

Non-parametric classification methods are powerful tools for modeling complex data without making strong assumptions about the underlying distribution. # Histogram Method

## Introduction
The Histogram method is a non-parametric technique used for density estimation and classification. It involves dividing the feature space into discrete bins and counting the number of data points that fall into each bin. This method is simple and intuitive, making it a popular choice for visualizing data distributions and performing basic classification tasks.

## How It Works
1. **Bin Division**: The feature space is divided into a grid of bins.
2. **Counting**: The number of data points in each bin is counted.
3. **Density Estimation**: The density of data points in each bin is estimated by dividing the count by the total number of data points and the bin volume.
4. **Classification**: For a new data point, the bin it falls into is determined, and the class label is assigned based on the majority class in that bin.

## Limitations
1. **Bin Size Sensitivity**: The choice of bin size can significantly affect the results. Too large bins may oversimplify the data, while too small bins may lead to overfitting.
2. **Curse of Dimensionality**: In high-dimensional spaces, the number of bins grows exponentially, making the method impractical.
3. **Boundary Issues**: Data points near the boundaries of bins may be misclassified due to the arbitrary bin divisions.
4. **Data Sparsity**: In cases of sparse data, many bins may be empty, leading to unreliable density estimates.

## Example in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X_class0 = np.random.randn(100, 2) + np.array([2, 2])
X_class1 = np.random.randn(100, 2) + np.array([-2, -2])
X = np.vstack((X_class0, X_class1))
y = np.array([0] * 100 + [1] * 100)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Define the bin edges
bin_edges = [np.linspace(-5, 5, 11), np.linspace(-5, 5, 11)]

# Create a histogram for each class
hist_class0, _ = np.histogramdd(X_train[y_train == 0], bins=bin_edges)
hist_class1, _ = np.histogramdd(X_train[y_train == 1], bins=bin_edges)

# Normalize the histograms
hist_class0 /= hist_class0.sum()
hist_class1 /= hist_class1.sum()

# Predict the class for each test point
def predict(X, hist_class0, hist_class1, bin_edges):
    bin_indices = [np.digitize(X[:, i], bin_edges[i]) - 1 for i in range(X.shape[1])]
    bin_indices = np.array(bin_indices).T
    predictions = []
    for idx in bin_indices:
        if np.all(idx >= 0) and np.all(idx < hist_class0.shape):
            p0 = hist_class0[tuple(idx)]
            p1 = hist_class1[tuple(idx)]
            predictions.append(0 if p0 > p1 else 1)
        else:
            predictions.append(0)  # Default to class 0 if out of bounds
    return np.array(predictions)

y_pred = predict(X_test, hist_class0, hist_class1, bin_edges)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Plot decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = predict(np.c_[xx.ravel(), yy.ravel()], hist_class0, hist_class1, bin_edges)
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.title('Histogram Method Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Plot 1D histograms for the first feature
plt.hist(X_train[y_train == 0][:, 0], bins=bin_edges[0], alpha=0.5, label='Class 0')
plt.hist(X_train[y_train == 1][:, 0], bins=bin_edges[0], alpha=0.5, label='Class 1')
plt.title('Histogram of Feature 1')
plt.xlabel('Feature 1')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Parzen Density Estimation

## Introduction
Parzen density estimation, also known as kernel density estimation (KDE), is a non-parametric method for estimating the probability density function of a random variable. Unlike parametric methods, which assume a specific form for the distribution, Parzen density estimation makes minimal assumptions about the underlying distribution of the data.

## Intuition
The basic idea behind Parzen density estimation is to estimate the density at a point by averaging the contributions of nearby data points. Each data point contributes to the density estimate based on its distance from the point of interest, with closer points contributing more.

## How It Works
1. **Kernel Function**: A kernel function $ K $ is chosen, which determines the shape of the contribution from each data point. Common choices include the Gaussian kernel, Epanechnikov kernel, and uniform kernel.
2. **Bandwidth**: A bandwidth parameter $ h $ is selected, which controls the width of the kernel and thus the smoothness of the density estimate.
3. **Density Estimation**: The density at a point $ x $ is estimated by summing the contributions from all data points, scaled by the kernel function and bandwidth.

### Formulas
Given a dataset $ $x_1, x_2, \ldots, x_n\} $, the Parzen density estimate at a point $ x $ is given by:
$ \hat{f}(x) = \frac{1}{n h^d} \sum_{i=1}^{n} K\left( \frac{x - x_i}{h} \right) $
where:
- $ n $ is the number of data points.
- $ h $ is the bandwidth.
- $ d $ is the dimensionality of the data.
- $ K $ is the kernel function.

### Common Kernel Functions
1. **Gaussian Kernel**:
$ K(u) = \frac{1}{\sqrt{2\pi}} \exp\left( -\frac{1}{2} u^2 \right) $
2. **Epanechnikov Kernel**:
$ K(u) = \frac{3}{4} (1 - u^2) \quad \text{for} \quad |u| \leq 1 $
3. **Uniform Kernel**:
$ K(u) = \frac{1}{2} \quad \text{for} \quad |u| \leq 1 $

## Best Use Cases
- **Unknown Distribution**: When the underlying distribution of the data is unknown or difficult to approximate with a parametric model.
- **Multimodal Distributions**: When the data has multiple modes (peaks) that are not well-captured by a single parametric distribution.
- **Small to Medium-Sized Datasets**: Works well with small to medium-sized datasets where the computational cost is manageable.

## Limitations
1. **Choice of Bandwidth**: The performance of the method is highly sensitive to the choice of bandwidth. Too small a bandwidth can lead to overfitting, while too large a bandwidth can lead to oversmoothing.
2. **Computational Complexity**: The method can be computationally expensive, especially for large datasets, as it requires evaluating the kernel function for each data point.
3. **Curse of Dimensionality**: In high-dimensional spaces, the method can suffer from the curse of dimensionality, where the density estimates become less reliable.

## Example in Python
Below is a detailed example of Parzen density estimation using a synthetic dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity

# Generate synthetic data
np.random.seed(0)
X = np.concatenate([np.random.normal(0, 1, 100), np.random.normal(5, 1, 100)])[:, np.newaxis]

# Define the bandwidth
bandwidth = 0.5

# Initialize the Kernel Density Estimator
kde = KernelDensity(kernel='gaussian', bandwidth=bandwidth)

# Fit the model
kde.fit(X)

# Generate points where we want to estimate the density
X_plot = np.linspace(-5, 10, 1000)[:, np.newaxis]

# Estimate the density
log_dens = kde.score_samples(X_plot)
dens = np.exp(log_dens)

# Plot the results
plt.figure(figsize=(10, 6))
plt.fill_between(X_plot[:, 0], dens, alpha=0.5)
plt.plot(X[:, 0], -0.01 - 0.02 * np.random.random(X.shape[0]), '+k')
plt.title('Parzen Density Estimation')
plt.xlabel('Data')
plt.ylabel('Density')
plt.show()
```

```python
#manual implementation of Parzen Density Estimation

import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(0)
X = np.concatenate([np.random.normal(0, 1, 100), np.random.normal(5, 1, 100)])[:, np.newaxis]

# Define the bandwidth
bandwidth = 0.5

# Manually implement the Kernel Density Estimation
def gaussian_kernel(x, xi, bandwidth):
    return np.exp(-0.5 * ((x - xi) / bandwidth) ** 2) / (bandwidth * np.sqrt(2 * np.pi))

def kde_manual(X, X_plot, bandwidth):
    dens = np.zeros(X_plot.shape[0])
    for i, x in enumerate(X_plot):
        dens[i] = np.sum(gaussian_kernel(x, X, bandwidth))
    return dens / X.shape[0]

# Generate points where we want to estimate the density
X_plot = np.linspace(-5, 10, 1000)[:, np.newaxis]

# Estimate the density manually
dens_manual = kde_manual(X, X_plot, bandwidth)

# Plot the results
plt.figure(figsize=(10, 6))
plt.fill_between(X_plot[:, 0], dens_manual, alpha=0.5)
plt.plot(X[:, 0], -0.01 - 0.02 * np.random.random(X.shape[0]), '+k')
plt.title('Manual Parzen Density Estimation')
plt.xlabel('Data')
plt.ylabel('Density')
plt.show()
```

## Parzen Density Estimation using a Uniform Kernel

### Intuition

Parzen density estimation is a non-parametric method to estimate the probability density function of a random variable. The idea is to place a kernel (a small window) around each data point and sum the contributions of all kernels to estimate the density at any given point.

For a uniform kernel, the kernel function $ K(r, h) $ is defined as:
$ K(r, h) = 
\begin{cases} 
0 & \text{if } |r| > h \\
\frac{1}{V} & \text{if } |r| \leq h 
\end{cases}
$
where $ V $ is the volume of the kernel, $ r $ is the distance from the point of interest to a data point, and $ h $ is the bandwidth (or window width).

This means that each data point contributes equally within a certain range (bandwidth) and not at all outside this range.

The bandwidth parameter controls the width of the kernel. It determines the range within which data points contribute to the density estimate. A larger bandwidth results in a smoother density estimate, while a smaller bandwidth can capture more details but may lead to overfitting.

### Steps to Create and Use the Classifier

1. **Define the Kernel Function**: Implement the uniform kernel function.
2. **Estimate Density**: For a test object $ z $, the density estimate $ p(z|h) $ is given by:
$ p(z|h) = \frac{1}{n} \sum_{i=1}^{n} K(||z - x_i||, h) $
where $ n $ is the number of data points, $ x_i $ are the data points, and $ ||z - x_i|| $ is the distance between  $ z $ and $ x_i $.

    - The density at a point ( x ) is estimated by summing the contributions from all data points within the bandwidth.

    - This step is done for each class separately to estimate the densities for each class, which gives P(x|y=k) for each class k.

3. **Classify**: Use the estimated densities to classify the test object.
    - After we estimate P(x|y=k) for each class k, we can use Bayes' theorem to calculate P(y=k|x) and classify the test object based on the class with the highest probability.

### Example in Python

Below is an example of implementing Parzen density estimation using a uniform kernel in Python. The example includes generating suitable data, plotting the kernels, and plotting the decision boundary.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the uniform kernel function
def uniform_kernel(r, h):
    V = (2 * h) ** 2  # Volume of the kernel in 2D
    if np.abs(r) > h:
        return 0
    else:
        return 1 / V

# Estimate density using Parzen window
def parzen_density_estimation(data, z, h):
    n = len(data)
    density = 0
    for x_i in data:
        density += uniform_kernel(np.linalg.norm(z - x_i), h)
    return density / n

# Generate sample data
np.random.seed(0)
data_class1 = np.random.randn(50, 2) * 0.5 + np.array([2, 2])
data_class2 = np.random.randn(50, 2) * 0.5 + np.array([-2, -2])

# Plot the data and kernels
h = 1.0
plt.figure(figsize=(10, 5))
plt.scatter(data_class1[:, 0], data_class1[:, 1], color='blue', label='Class 1')
plt.scatter(data_class2[:, 0], data_class2[:, 1], color='red', label='Class 2')

# Draw kernels as circles
for point in data_class1:
    circle = plt.Circle(point, h, color='blue', fill=False, linestyle='dotted')
    plt.gca().add_artist(circle)
for point in data_class2:
    circle = plt.Circle(point, h, color='red', fill=False, linestyle='dotted')
    plt.gca().add_artist(circle)

plt.legend()
plt.title('Data Points and Kernels')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()

# Plot decision boundary
x_min, x_max = -4, 4
y_min, y_max = -4, 4
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Compute density estimates for each class
density_class1 = np.array([parzen_density_estimation(data_class1, z, h) for z in grid_points])
density_class2 = np.array([parzen_density_estimation(data_class2, z, h) for z in grid_points])

# Decision boundary: where densities are equal
decision_boundary = density_class1 - density_class2
decision_boundary = decision_boundary.reshape(xx.shape)

plt.figure(figsize=(10, 5))
plt.contourf(xx, yy, decision_boundary, levels=[-np.inf, 0, np.inf], colors=['red', 'blue'], alpha=0.3)
plt.scatter(data_class1[:, 0], data_class1[:, 1], color='blue', label='Class 1')
plt.scatter(data_class2[:, 0], data_class2[:, 1], color='red', label='Class 2')
plt.legend()
plt.title('Decision Boundary')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
```

## Parzen Density Estimation with Gaussian Kernel for single 1D data
### Kernel Formula

The kernel function $ K(x) $ is a non-negative function that integrates to one. For the Gaussian kernel, the formula is:

$$
K(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} x^2}
$$

This is the probability density function of the standard normal distribution.

### Density Estimation Formula

Parzen density estimation, also known as Kernel Density Estimation (KDE), estimates the probability density function $ P(x) $ of a random variable $ x $. The formula for KDE using a Gaussian kernel is:

$$
\hat{f}(x) = \frac{1}{n h} \sum_{i=1}^{n} K\left(\frac{x - x_i}{h}\right)
$$

where:
- $ \hat{f}(x) $ is the estimated density at point $ x $.
- $ n $ is the number of data points.
- $ h $ is the bandwidth (smoothing parameter).
- $ x_i $ are the data points.
- $ K $ is the kernel function.

### What Density Do We Estimate?

In KDE, we estimate the probability density function $ P(x) $. This is the density of the data points themselves, not conditional on any other variable. So, it is $ P(x) $, not $ P(x|y) $.

### Finding the Posterior Probability

To find the posterior probability $ P(y|x) $, you can use Bayes' theorem:

$$
P(y|x) = \frac{P(x|y) P(y)}{P(x)}
$$

Here:
- $ P(x|y) $ is the likelihood, which can be estimated using KDE for each class $ y $.
- $ P(y) $ is the prior probability of class $ y $.
- $ P(x) $ is the marginal probability, which can be computed as:

$$
P(x) = \sum_{y} P(x|y) P(y)
$$

### Means and Variance for Gaussian Kernel

For the Gaussian kernel, the mean $ \mu $ and variance $ \sigma^2 $ are parameters of the normal distribution. In the context of KDE, the mean is typically the data point $ x_i $, and the variance is related to the bandwidth $ h $:

$$
\sigma^2 = h^2
$$

### Comparison: Gaussian Kernel vs. Uniform Kernel

- **Gaussian Kernel**: Smooth and continuous, providing a smooth estimate of the density. It has infinite support, meaning it affects the density estimate at all points, though the influence decreases with distance from the data point.

  $$ K(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} x^2} $$

- **Uniform Kernel**: Simpler and has finite support. It assigns equal weight to all points within a certain distance (bandwidth) and zero weight outside that distance.

  $$ K(x) = \begin{cases} 
  \frac{1}{2} & \text{if } |x| \leq 1 \\
  0 & \text{otherwise}
  \end{cases} $$

### Summary of Formulas

1. **Gaussian Kernel**: $ K(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{1}{2} x^2} $
2. **Density Estimation**: $ \hat{f}(x) = \frac{1}{n h} \sum_{i=1}^{n} K\left(\frac{x - x_i}{h}\right) $
3. **Posterior Probability**: $ P(y|x) = \frac{P(x|y) P(y)}{P(x)} $
4. **Uniform Kernel**: $ K(x) = \begin{cases} \frac{1}{2} & \text{if } |x| \leq 1 \\ 0 & \text{otherwise} \end{cases} $

### Density Estimation Formula from Your Book

The formula you provided is:

$$
p(X|w_i) = \frac{1}{n_i} \sum_{j=1}^{n_i} N(x | (x_j)^i, hI)
$$

Here's what each part means:

- $ p(X|w_i) $: This is the estimated density of $ X $ given the class $ w_i $.
- $ n_i $: The number of data points in class $ w_i $.
- $ N(x | (x_j)^i, hI) $: This is the Gaussian (normal) distribution with mean $ (x_j)^i $ and covariance matrix $ hI $.

### Gaussian Distribution

The Gaussian distribution $ N(x | \mu, \Sigma) $ is given by:

$$
N(x | \mu, \Sigma) = \frac{1}{(2\pi)^{d/2} |\Sigma|^{1/2}} \exp\left(-\frac{1}{2} (x - \mu)^T \Sigma^{-1} (x - \mu)\right)
$$

In your formula:
- $ \mu = (x_j)^i $: The mean is the data point $ (x_j)^i $ from class $ w_i $.
- $ \Sigma = hI $: The covariance matrix is $ hI $, where $ h $ is the bandwidth and $ I $ is the identity matrix.

### Putting It All Together

The formula $ p(X|w_i) = \frac{1}{n_i} \sum_{j=1}^{n_i} N(x | (x_j)^i, hI) $ means that the density estimate for $ X $ given class $ w_i $ is the average of the Gaussian densities centered at each data point $ (x_j)^i $ in class $ w_i $, with covariance matrix $ hI $.

### Comparison to Previous Formula

The main difference between this and the previous KDE formula is that this one is specifically for estimating the density given a class $ w_i $, and it uses the Gaussian distribution explicitly in the formula.

### Summary

- **Kernel Density Estimation (KDE)**: Estimates the overall density $ P(x) $ using a sum of kernel functions.
- **Class-Conditional Density Estimation**: Estimates the density $ p(X|w_i) $ for a specific class $ w_i $ using a sum of Gaussian distributions centered at the data points of that class.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Generate synthetic data
np.random.seed(0)
X_class0 = np.random.normal(0, 1, 100)
X_class1 = np.random.normal(5, 1, 100)
X = np.concatenate([X_class0, X_class1])
y = np.array([0] * 100 + [1] * 100)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Define the Gaussian kernel function
def gaussian_kernel(x, xi, bandwidth):
    return (1 / (np.sqrt(2 * np.pi) * bandwidth)) * np.exp(-0.5 * ((x - xi) / bandwidth) ** 2)

# Estimate density using Parzen window
def parzen_density_estimation(data, z, bandwidth):
    n = len(data)
    return np.sum([gaussian_kernel(z, xi, bandwidth) for xi in data]) / n

# Define the bandwidth
bandwidth = 0.5

# Generate points where we want to estimate the density
X_plot = np.linspace(-3, 8, 1000)

# Estimate the density for each class
density_class0 = np.array([parzen_density_estimation(X_train[y_train == 0], z, bandwidth) for z in X_plot])
density_class1 = np.array([parzen_density_estimation(X_train[y_train == 1], z, bandwidth) for z in X_plot])

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(X_plot, density_class0, label='Class 0 Density', color='blue')
plt.plot(X_plot, density_class1, label='Class 1 Density', color='red')
plt.fill_between(X_plot, density_class0, alpha=0.3, color='blue')
plt.fill_between(X_plot, density_class1, alpha=0.3, color='red')
plt.scatter(X_train[y_train == 0], np.zeros_like(X_train[y_train == 0]), color='blue', marker='|', label='Class 0 Data')
plt.scatter(X_train[y_train == 1], np.zeros_like(X_train[y_train == 1]), color='red', marker='|', label='Class 1 Data')
plt.title('Parzen Density Estimation with Gaussian Kernel')
plt.xlabel('Data')
plt.ylabel('Density')
plt.legend()
plt.show()

# Plot decision boundary
decision_boundary = density_class0 - density_class1
plt.figure(figsize=(10, 6))
plt.plot(X_plot, decision_boundary, label='Decision Boundary', color='green')
plt.axhline(0, color='gray', linestyle='--')
plt.scatter(X_train[y_train == 0], np.zeros_like(X_train[y_train == 0]), color='blue', marker='|', label='Class 0 Data')
plt.scatter(X_train[y_train == 1], np.zeros_like(X_train[y_train == 1]), color='red', marker='|', label='Class 1 Data')
plt.title('Decision Boundary')
plt.xlabel('Data')
plt.ylabel('Density Difference')
plt.legend()
plt.show()
```

## Parzen Density Estimation with Multivariate Gaussian Kernel

$$
K(x) = \frac{1}{(2\pi)^{d/2} |\Sigma|^{1/2}} \exp\left(-\frac{1}{2} (x - \mu)^T \Sigma^{-1} (x - \mu)\right)
$$

where:
- $ d $ is the dimensionality of the data.
- $ \mu $ is the mean vector.
- $ \Sigma $ is the covariance matrix.

The formula you provided is:

p(X|w_i) = \frac{1}{n_i} \sum_{j=1}^{n_i} N(x | (x_j)^i, hI)

- $ n_i $: The number of data points in class $ w_i $.
- $ N(x | (x_j)^i, hI) $: The multivariate Gaussian distribution with mean $ (x_j)^i $ and covariance matrix $ hI $.

### Multivariate Gaussian Distribution

The multivariate Gaussian distribution $ N(x | \mu, \Sigma) $ is given by:

$$
N(x | \mu, \Sigma) = \frac{1}{(2\pi)^{d/2} |\Sigma|^{1/2}} \exp\left(-\frac{1}{2} (x - \mu)^T \Sigma^{-1} (x - \mu)\right)
$$

In your formula:
- $ \mu = (x_j)^i $: The mean is the data point $ (x_j)^i $ from class $ w_i $.

### Putting It All Together

The formula $ p(X|w_i) = \frac{1}{n_i} \sum_{j=1}^{n_i} N(x | (x_j)^i, hI) $ means that the density estimate for $ X $ given class $ w_i $ is the average of the multivariate Gaussian densities centered at each data point $ (x_j)^i $ in class $ w_i $, with covariance matrix $ hI $.

### Summary

- **1D Gaussian Kernel**: Used for univariate data.
- **Multivariate Gaussian Kernel**: Used for multivariate data, with the kernel function incorporating the covariance matrix $ \Sigma $.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Generate synthetic 2D data
np.random.seed(0)
X_class0 = np.random.randn(100, 2) + np.array([2, 2])
X_class1 = np.random.randn(100, 2) + np.array([-2, -2])
X = np.vstack((X_class0, X_class1))
y = np.array([0] * 100 + [1] * 100)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Define the Gaussian kernel function for 2D data
def gaussian_kernel(x, xi, bandwidth):
    return np.exp(-np.linalg.norm(x - xi)**2 / (2 * bandwidth**2)) / (2 * np.pi * bandwidth**2)

# Estimate density using Parzen window
def parzen_density_estimation(data, z, bandwidth):
    n = data.shape[0]
    return np.sum([gaussian_kernel(z, data[i], bandwidth) for i in range(n)]) / n

# Define the bandwidth
bandwidth = 0.5

# Generate points where we want to estimate the density
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# Estimate the density for each class
density_class0 = np.array([parzen_density_estimation(X_train[y_train == 0], z, bandwidth) for z in grid_points])
density_class1 = np.array([parzen_density_estimation(X_train[y_train == 1], z, bandwidth) for z in grid_points])

# Reshape the density estimates
density_class0 = density_class0.reshape(xx.shape)
density_class1 = density_class1.reshape(xx.shape)

# Plot the results
plt.figure(figsize=(12, 6))
plt.contourf(xx, yy, density_class0, alpha=0.5, cmap='Blues')
plt.contourf(xx, yy, density_class1, alpha=0.5, cmap='Reds')
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='blue', label='Class 0')
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='red', label='Class 1')
plt.title('Parzen Density Estimation with Gaussian Kernel')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Plot decision boundary
decision_boundary = density_class0 - density_class1
plt.figure(figsize=(12, 6))
plt.contourf(xx, yy, decision_boundary, levels=[-np.inf, 0, np.inf], colors=['red', 'blue'], alpha=0.3)
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='blue', label='Class 0')
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='red', label='Class 1')
plt.title('Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

# Parzen Width Parameter Optimization

## Importance of Choosing the Parzen Width Parameter Carefully

The Parzen width parameter, often denoted as $ h $, is crucial in kernel density estimation (KDE). It determines the smoothness of the estimated density function. Choosing $ h $ carefully is important because:

1. **Bias-Variance Trade-off**:
- **Small $ h $**: Leads to a high-variance, low-bias estimate. The density estimate will be very sensitive to the data, resulting in a "noisy" estimate.
- **Large $ h $**: Leads to a low-variance, high-bias estimate. The density estimate will be overly smooth, potentially missing important features of the data.

2. **Overfitting and Underfitting**:
- **Overfitting**: A very small $ h $ can cause the model to fit the noise in the data.
- **Underfitting**: A very large $ h $ can cause the model to miss important patterns in the data.

## Optimizing the Parzen Width Parameter

### 1. Average k-Nearest Neighbor Distance

One method to optimize $ h $ is to use the average distance to the k-nearest neighbors. This method adapts $ h $ based on the local density of the data points.

### 2. Heuristics

#### Heuristic 1: Silverman's Rule of Thumb

$ h = \sigma \left( \frac{4}{p+2} \right)^{\frac{1}{p+4}} n^{-\frac{1}{p+4}} $

- $ \sigma $: Standard deviation of the data.
- $ p $: Number of dimensions.
- $ n $: Number of data points.

This heuristic provides a good starting point for $ h $ by considering the dimensionality and spread of the data.

#### Heuristic 2: Variance-Based Method

$ \sigma^2 = \frac{1}{P} \sum_{i=1}^{P} s_{ii} $

- $ P $: Number of dimensions.
- $ s_{ii} $: Variance of the $ i $-th dimension.

This method calculates the average variance across all dimensions to determine $ h $.

### 3. Likelihood Using Leave-One-Out Cross-Validation

Leave-one-out cross-validation (LOO-CV) can be used to optimize $ h $ by maximizing the likelihood of the data given the KDE model. The idea is to leave one data point out, estimate the density using the remaining points, and evaluate the likelihood of the left-out point. This process is repeated for all data points, and the average likelihood is maximized.

## Intuition Behind the Methods

- **Average k-Nearest Neighbor Distance**: This method adapts $ h $ based on the local density of data points, ensuring that the estimate is neither too smooth nor too noisy.
- **Silverman's Rule of Thumb**: This heuristic balances the bias-variance trade-off by considering the dimensionality and spread of the data.
- **Variance-Based Method**: This method uses the average variance across dimensions to determine $ h $, ensuring that the estimate is appropriate for the data's spread.
- **Leave-One-Out Cross-Validation**: This method ensures that the chosen $ h $ maximizes the likelihood of the data, providing a robust estimate.

By carefully choosing and optimizing the Parzen width parameter, we can achieve a more accurate and reliable density estimate.



# Nearest Neighbor Classification

## What is Nearest Neighbor Classification?

Nearest Neighbor Classification is a type of instance-based learning where the classification of a new data point is determined based on the classes of its nearest neighbors in the training data. The most common form is the k-Nearest Neighbors (k-NN) algorithm.

### k-Nearest Neighbors (k-NN) Algorithm

1. **Training Phase**:
- The k-NN algorithm does not have an explicit training phase. It simply stores the training data.

2. **Classification Phase**:
- For a new data point $ x $, the algorithm finds the $ k $ training points that are closest to $ x $ (using a distance metric like Euclidean distance).
- The class of $ x $ is determined by the majority class among its $ k $ nearest neighbors.

## k-NN Density Estimate

The k-NN density estimate is a way to estimate the probability density function $ p(x) $ using the k-nearest neighbors. The formula is:

$ p(x) = \frac{k}{nV_k} $

- $ k $: Number of nearest neighbors.
- $ n $: Total number of data points.
- $ V_k $: Volume of the region containing the $ k $ nearest neighbors.

### Estimating $ p(x) $ or $ p(x|y) $

- **For all data**: If you are estimating the overall density $ p(x) $, you consider all data points regardless of their class.
    - **Per class**: If you are estimating the class-conditional density $ p(x|y) $, you consider only the data points belonging to class $ y $.

## Explanation of $ r $ and $ k $

- **$ r $**: The radius of the region containing the $ k $ nearest neighbors. It is the distance from the query point $ x $ to its $ k $-th nearest neighbor.
- **$ k $**: The number of nearest neighbors considered for the density estimate or classification. It is a hyperparameter that needs to be chosen carefully.

## Choice of $ k $

The choice of $ k $ is crucial for the performance of the k-NN algorithm. A small $ k $ can lead to a noisy estimate (high variance), while a large $ k $ can lead to a smooth estimate (high bias). Cross-validation is often used to choose the optimal $ k $.

## Python Example

Here's a Python example using 2D data to visualize the results and draw the decision boundary:



```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier
from matplotlib.colors import ListedColormap

# Generate synthetic data
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, random_state=42)
h = .02  # step size in the mesh

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#0000FF'])

# Fit the k-NN classifier
k = 5
clf = KNeighborsClassifier(n_neighbors=k)
clf.fit(X, y)

# Plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=cmap_light)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor='k', s=20)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title(f"k-NN classification (k = {k})")
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
