# Determinant and Inverse of a Matrix

## Determinant of a Matrix

The determinant of a matrix is a scalar value that is a function of the entries of a square matrix. It provides important properties of the matrix and can be used to determine whether a matrix is invertible.

### Intuition
- **Geometric Interpretation**: The determinant of a 2x2 matrix can be interpreted as the area of the parallelogram formed by its column vectors. For higher dimensions, it represents the volume of the parallelepiped.
- **Properties**:
  - If the determinant is zero, the matrix is singular (non-invertible).
  - The sign of the determinant can indicate the orientation of the transformation described by the matrix.

### Calculation
For a 2x2 matrix:
$ A = \begin{pmatrix} a & b \\ c & d \end{pmatrix} $
The determinant is calculated as:
$ \text{det}(A) = ad - bc $

For larger matrices, the determinant is calculated using more complex methods such as cofactor expansion.

## Inverse of a Matrix

The inverse of a matrix $ A $ is another matrix $ A^{-1} $ such that:
$ A A^{-1} = A^{-1} A = I $
where $ I $ is the identity matrix.

### Intuition
- **Geometric Interpretation**: The inverse of a matrix represents a transformation that "undoes" the transformation represented by the original matrix.
- **Properties**:
  - A matrix is invertible if and only if its determinant is non-zero.
  - The inverse of a product of matrices is the product of their inverses in reverse order.

### Calculation
For a 2x2 matrix:
$ A = \begin{pmatrix} a & b \\ c & d \end{pmatrix} $
The inverse is calculated as:
$ A^{-1} = \frac{1}{\text{det}(A)} \begin{pmatrix} d & -b \\ -c & a \end{pmatrix} $
provided that $ \text{det}(A) \neq 0 $.

## Example in Python


```python
import numpy as np

# Define a 2x2 matrix
A = np.array([[4, 7], [2, 6]])

# Calculate the determinant
det_A = np.linalg.det(A)
print(f"Determinant of A: {det_A}")

# Calculate the inverse
if det_A != 0:
    inv_A = np.linalg.inv(A)
    print(f"Inverse of A:\n{inv_A}")
else:
    print("Matrix A is singular and does not have an inverse.")
```

## Use in Gaussian Distribution

In the context of the multivariate Gaussian distribution, the determinant and inverse of the covariance matrix $ \Sigma $ play crucial roles.

### Determinant
- **Volume Scaling**: The determinant of $ \Sigma $ scales the volume of the distribution. A larger determinant indicates a more spread out distribution, while a smaller determinant indicates a more concentrated distribution.

### Inverse
- **Mahalanobis Distance**: The inverse of $ \Sigma $ is used to compute the Mahalanobis distance, which measures the distance between a point and the mean of the distribution, taking into account the correlations between variables.

### Probability Density Function (PDF)
The PDF of a multivariate Gaussian distribution is given by:
$ f(\mathbf{x}) = \frac{1}{(2\pi)^{k/2} |\Sigma|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x} - \mu)^T \Sigma^{-1} (\mathbf{x} - \mu)\right) $

where:
- $ \mathbf{x} $ is a $ k $-dimensional random vector.
- $ \mu $ is the mean vector.
- $ \Sigma $ is the covariance matrix.
- $ |\Sigma| $ is the determinant of the covariance matrix.
- $ \Sigma^{-1} $ is the inverse of the covariance matrix.

The determinant $ |\Sigma| $ normalizes the distribution, while the inverse $ \Sigma^{-1} $ adjusts for the correlations between variables.

# Gaussian Distribution

## Introduction
The Gaussian distribution, also known as the normal distribution, is a continuous probability distribution that is symmetrical around its mean. It is characterized by its bell-shaped curve.

## Properties
1. **Mean** ($\mu$) : The central value around which the distribution is symmetrical.
2. **Standard Deviation** ($\sigma$): Measures the spread of the distribution. A smaller standard deviation indicates that the data points are closer to the mean, while a larger standard deviation indicates that the data points are spread out over a wider range of values.
3. **Variance** ($\sigma^2$): The square of the standard deviation.

## Probability Density Function (PDF)
The PDF of a Gaussian distribution is given by:

$ f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}} $

## Example

```python
#Let's generate a Gaussian distribution using Python and visualize it.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate data
x = np.linspace(-5, 5, 1000)
y = norm.pdf(x, mu, sigma)

# Plot
plt.plot(x, y)
plt.title('Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```

# Multivariate Gaussian Distribution

## Introduction
The multivariate Gaussian distribution is a generalization of the one-dimensional (univariate) Gaussian distribution to higher dimensions. It describes the joint distribution of a set of random variables that may be correlated with each other.

## Intuition
- **Univariate vs. Multivariate**: While the univariate Gaussian distribution describes a single random variable, the multivariate Gaussian distribution describes multiple random variables simultaneously.
- **Correlation**: In the multivariate case, the variables can be correlated, meaning that the value of one variable can give information about the value of another.

## Properties
1. **Mean Vector ($\mu$)**: A vector of means for each variable.
2. **Covariance Matrix ($\Sigma$)**: A matrix that describes the variance of each variable along the diagonal and the covariance between variables off the diagonal.

## Probability Density Function (PDF)
The PDF of a multivariate Gaussian distribution is given by:

$ f(\mathbf{x}) = \frac{1}{(2\pi)^{k/2} |\Sigma|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x} - \mu)^T \Sigma^{-1} (\mathbf{x} - \mu)\right) $
where:
- $\mathbf{x}$ is a $k$-dimensional random vector, representing the variables.
    - k means the number of variables.
    - $k=2$ represents a bivariate distribution.
- $\mu$ is the mean vector.
- $\Sigma$ is the covariance matrix.
- $|\Sigma|$ is the determinant of the covariance matrix.
- $\Sigma^{-1}$ is the inverse of the covariance matrix.

## Example
Let's generate a multivariate Gaussian distribution using Python and visualize it.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Parameters
mu = np.array([0, 0])  # Mean vector
sigma = np.array([[1, 0.5], [0.5, 1]])  # Covariance matrix

# Generate grid
x, y = np.mgrid[-3:3:.01, -3:3:.01]
pos = np.dstack((x, y))

# Multivariate normal distribution
rv = multivariate_normal(mu, sigma)
z = rv.pdf(pos)

# Plot
plt.contourf(x, y, z, cmap='viridis')
plt.title('Multivariate Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
plt.show()
```

# Parametric Classifiers

## Introduction
Parametric classifiers are a type of classification algorithm that make assumptions about the underlying distribution of the data. These classifiers use a fixed number of parameters to describe the data, which simplifies the learning process and often leads to faster and more efficient algorithms.

## Key Characteristics
1. **Assumptions**: They assume a specific form for the distribution of the data (e.g., Gaussian distribution).
2. **Fixed Parameters**: The number of parameters is fixed and does not grow with the size of the dataset.
3. **Efficiency**: They are generally computationally efficient due to the fixed number of parameters.
4. **Examples**: Common examples include Linear Discriminant Analysis (LDA), Quadratic Discriminant Analysis (QDA), and Gaussian Naive Bayes.

# Gaussian-based Classifiers

## Introduction
Gaussian-based classifiers are a subset of parametric classifiers that assume the data follows a Gaussian (normal) distribution. These classifiers are particularly useful when the data is continuous and can be well-approximated by a Gaussian distribution.

## Gaussian Naive Bayes
Gaussian Naive Bayes is a probabilistic classifier based on Bayes' theorem with the assumption that the features are normally distributed.

### Assumptions
1. **Feature Independence**: It assumes that the features are independent given the class label.
2. **Gaussian Distribution**: It assumes that the features follow a Gaussian distribution.

### Probability Density Function (PDF)
The PDF of a Gaussian distribution for a feature $ x_i $ is given by:
$ f(x_i) = \frac{1}{\sqrt{2\pi\sigma_i^2}} \exp\left(-\frac{(x_i - \mu_i)^2}{2\sigma_i^2}\right) $
where:
- $ \mu_i $ is the mean of the feature.
- $ \sigma_i^2 $ is the variance of the feature.

### Classification Rule
The classifier assigns a class label $ y $ to a new data point $ \mathbf{x} $ by maximizing the posterior probability:
$ y = \arg\max_y P(y) \prod_{i=1}^n f(x_i | y) $
where:
- $ P(y) $ is the prior probability of class $ y $.
- $ f(x_i | y) $ is the likelihood of feature $ x_i $ given class $ y $.

## Example in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X = np.random.randn(100, 2)
y = np.array([0 if x[0] + x[1] < 0 else 1 for x in X])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Initialize the Gaussian Naive Bayes classifier
gnb = GaussianNB()

# Fit the model
gnb.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = gnb.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Plot the decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = gnb.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.title('Gaussian Naive Bayes Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

# Quadratic Discriminant Analysis (QDA)

## Intuition

Quadratic Discriminant Analysis (QDA) is a classification algorithm that assumes that the data from each class is drawn from a Gaussian distribution. Unlike Linear Discriminant Analysis (LDA), which assumes that each class shares the same covariance matrix, QDA allows each class to have its own covariance matrix. This flexibility allows QDA to model more complex decision boundaries.

### Discriminant Function

This function combines the likelihood of the data point belonging to the class, the prior probability of the class, and the covariance structure of the class.

The discriminant function for QDA is given by:
$ \delta_k(\mathbf{x}) = -\frac{1}{2} \log |\Sigma_k| - \frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma_k^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $
where:

- $ \delta_k(\mathbf{x}) $ is the discriminant function for class $ k $.
- $ \mathbf{x} $ is the input data.
- $ \mu_k $ is the mean vector of class $ k $.
- $ \Sigma_k $ is the covariance matrix of class $ k $.
- P(y = k) is the **prior probability** of class $ k $.

#### General Form
$ f(\mathbf{x}) = \mathbf{x}^T W \mathbf{x} + \mathbf{w}^T \mathbf{x} + w_0 $
- Quadratic term: $ \mathbf{x}^T W \mathbf{x} $
    - This term captures the curvature of the decision boundary.
- Linear term: $ \mathbf{w}^T \mathbf{x} $
    - This term captures the influence of the input features.
    - This term captures the linear relationship between the features and the class.
- Constant term: $ w_0 $
    - This term adjusts the decision boundary based on the class priors and the covariance structure.
 

### Key Characteristics
1. **Class-Specific Covariance Matrices**: Each class has its own covariance matrix, allowing for more flexible decision boundaries.
2. **Quadratic Decision Boundaries**: The decision boundaries between classes are quadratic (parabolic) rather than linear.
3. **Gaussian Assumption**: Assumes that the data for each class follows a Gaussian distribution.

## Best Fit Data

QDA is best suited for:
- **Data with Non-Linear Boundaries**: When the decision boundaries between classes are not linear.
- **Sufficient Data**: Requires a relatively large amount of data to accurately estimate the covariance matrices for each class.
- **Continuous Features**: Works well with continuous data that can be approximated by a Gaussian distribution.

## Relevance of Probabilities in QDA

In the context of classification, particularly with Quadratic Discriminant Analysis (QDA), the probabilities $ P(x) $, $ P(y) $, $ P(x|y) $, and $ P(y|x) $ play specific roles:

1. **$ P(x) $ - Marginal Probability of $ x $**:
   - **Relevance**: This is the overall probability of observing the feature vector $ x $ in the dataset.
   - **Known or Estimated**: It is generally not directly estimated in QDA. Instead, it is implicitly considered when normalizing probabilities.
   - **Calculation**: Not directly calculated in QDA.

2. **$ P(y) $ - Prior Probability of Class $ y $**:
   - **Relevance**: This is the probability of a data point belonging to class $ y $ before observing the feature vector $ x $.
   - **Known or Estimated**: Estimated from the training data as the proportion of data points in each class.
   - **Calculation**: Estimated during the training phase.

3. **$ P(x|y) $ - Likelihood of $ x $ Given Class $ y $**:
   - **Relevance**: This is the probability of observing the feature vector $ x $ given that it belongs to class $ y $. In QDA, this is modeled using a Gaussian distribution with class-specific mean and covariance.
   - **Known or Estimated**: Estimated from the training data by fitting a Gaussian distribution to the data points in each class.
   - **Calculation**: Estimated during the training phase.

4. **$ P(y|x) $ - Posterior Probability of Class $ y $ Given $ x $**:
   - **Relevance**: This is the probability of a data point belonging to class $ y $ after observing the feature vector $ x $. This is what we aim to calculate when classifying a new data point.
   - **Known or Estimated**: Calculated using Bayes' theorem during the classification phase.
   - **Calculation**: Calculated when a new $ x $ needs to be classified.

### Summary of Estimation and Calculation

- **Estimated from Training Data**:
  - $ P(y) $
  - $ P(x|y) $ (through the mean vector $ \mu_k $ and covariance matrix $ \Sigma_k $)

- **Calculated During Classification**:
  - $ P(y|x) $

### Bayes' Theorem in Classification

To classify a new data point $ x $, we use Bayes' theorem:
$ P(y|x) = \frac{P(x|y) \cdot P(y)}{P(x)} $

In practice, we compare the discriminant functions for each class:
$ \delta_k(\mathbf{x}) = -\frac{1}{2} \log |\Sigma_k| - \frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma_k^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $

### Why Use the Discriminant Function?
Avoiding $ P(x) $: Directly calculating $ P(x) $ can be complex and unnecessary for classification. Since $ P(x) $ is the same for all classes, it cancels out when comparing classes.
Computational Efficiency: The discriminant function combines the necessary terms in a way that is computationally efficient and avoids the need to explicitly calculate $P(x)$.



The class with the highest discriminant function value is chosen as the predicted class.

Quadratic Discriminant Analysis (QDA) is a powerful classification algorithm that models class-specific covariance matrices, allowing for quadratic decision boundaries. It is best suited for data with non-linear boundaries and requires a sufficient amount of data to accurately estimate the parameters. Understanding the underlying Gaussian assumptions and the discriminant function is crucial for effectively applying QDA to classification problems.

## How It Works

1. **Estimate Parameters**: For each class, estimate the mean vector $ \mu_k $ and the covariance matrix $ \Sigma_k $.
2. **Compute Discriminant Function**: For a new data point $ \mathbf{x} $, compute the discriminant function for each class.
3. **Assign Class**: Assign the class label that maximizes the discriminant function.


### Quadratic Decision Boundaries

The term "quadratic" in QDA refers to the fact that the decision boundaries between classes are quadratic functions of the input features. This is due to the class-specific covariance matrices, which result in a quadratic form in the discriminant function.

## Formulas

### Mean Vector and Covariance Matrix Estimation
For each class $ k $:
- Mean vector $ \mu_k $:
$ \mu_k = \frac{1}{N_k} \sum_{i \in C_k} \mathbf{x}_i $
where $ N_k $ is the number of samples in class $ k $ and $ \mathbf{x}_i $ are the samples in class $ k $.

- Covariance matrix $ \Sigma_k $:
$ \Sigma_k = \frac{1}{N_k - 1} \sum_{i \in C_k} (\mathbf{x}_i - \mu_k)(\mathbf{x}_i - \mu_k)^T $

### Discriminant Function
The discriminant function for class $ k $ is given by:
$ \delta_k(\mathbf{x}) = -\frac{1}{2} \log |\Sigma_k| - \frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma_k^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $
where:
- $ |\Sigma_k| $ is the determinant of the covariance matrix for class $ k $.
- $ \Sigma_k^{-1} $ is the inverse of the covariance matrix for class $ k $.
- $ P(y = k) $ is the prior probability of class $ k $.

### Classification Rule
Assign the class label $ y $ to a new data point $ \mathbf{x} $ by maximizing the discriminant function:
$ y = \arg\max_k \delta_k(\mathbf{x}) $

## Example in Python

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X = np.random.randn(100, 2)
y = np.array([0 if x[0] + x[1] < 0 else 1 for x in X])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Initialize the Gaussian Naive Bayes classifier
gnb = GaussianNB()

# Fit the model
gnb.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = gnb.predict(X_test)

# Calculate the accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Plotting decision boundary in 2D
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
Z = gnb.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
plt.title('2D Decision Boundary')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Plotting decision boundary in 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], y, c=y, marker='o')

# Create a meshgrid for 3D plot
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))
zz = gnb.predict(np.c_[xx.ravel(), yy.ravel()])
zz = zz.reshape(xx.shape)

# Plot the surface
ax.plot_surface(xx, yy, zz, alpha=0.3, rstride=100, cstride=100)

ax.set_title('3D Decision Boundary')
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Class')
plt.show()
```

# Linear Discriminant Analysis (LDA) vs. Quadratic Discriminant Analysis (QDA)

## Introduction to Linear Discriminant Analysis (LDA)
Linear Discriminant Analysis (LDA) is a classification algorithm that assumes that the data from each class is drawn from a Gaussian distribution with the same covariance matrix. This results in linear decision boundaries between classes.

### Key Characteristics of LDA
1. **Shared Covariance Matrix**: All classes share the same covariance matrix.
2. **Linear Decision Boundaries**: The decision boundaries between classes are linear.
3. **Gaussian Assumption**: Assumes that the data for each class follows a Gaussian distribution.

### Discriminant Function for LDA
The discriminant function for LDA is given by:
    $ \delta_k(\mathbf{x}) = \mathbf{x}^T \Sigma^{-1} \mu_k - \frac{1}{2} \mu_k^T \Sigma^{-1} \mu_k + \log P(y = k) $
where:
- $ \delta_k(\mathbf{x}) $ is the discriminant function for class $ k $.
- $ \mathbf{x} $ is the input data.
- $ \mu_k $ is the mean vector of class $ k $.
- $ \Sigma $ is the shared covariance matrix.
- $ P(y = k) $ is the prior probability of class $ k $.

## Introduction to Quadratic Discriminant Analysis (QDA)
Quadratic Discriminant Analysis (QDA) is a classification algorithm that assumes that the data from each class is drawn from a Gaussian distribution. Unlike LDA, QDA allows each class to have its own covariance matrix. This flexibility allows QDA to model more complex decision boundaries.

### Key Characteristics of QDA
1. **Class-Specific Covariance Matrices**: Each class has its own covariance matrix, allowing for more flexible decision boundaries.
2. **Quadratic Decision Boundaries**: The decision boundaries between classes are quadratic (parabolic) rather than linear.
3. **Gaussian Assumption**: Assumes that the data for each class follows a Gaussian distribution.

## Intuition Behind Averaging Covariance Matrices
When there is insufficient data, the covariance matrix for a class may become singular (non-invertible). This can happen if one of the variances is zero, indicating that there is no variability in one of the features for that class. To address this issue, we can average the covariance matrices across all classes, effectively pooling the data to create a more stable estimate of the covariance.

### Is This Still QDA?
No, this approach is more akin to Linear Discriminant Analysis (LDA). In LDA, we assume that all classes share the same covariance matrix, which is estimated by averaging the covariance matrices of all classes. This results in linear decision boundaries rather than quadratic ones.

### Using the Averaged Covariance Matrix
When using the averaged covariance matrix, the same covariance matrix is used for all class discriminant functions. This means that the only class-specific parameters are the means.

### Formulas

#### Averaged Covariance Matrix
The averaged covariance matrix $ \Sigma $ is calculated as:
$ \Sigma = \frac{1}{K} \sum_{k=1}^{K} \Sigma_k $
where $ K $ is the number of classes and $ \Sigma_k $ is the covariance matrix for class $ k $.

#### Discriminant Function
The discriminant function for class $ k $ in this case is:
    $ \delta_k(\mathbf{x}) = -\frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $
where:
- $ \mu_k $ is the mean vector for class $ k $.
- $ \Sigma $ is the averaged covariance matrix.
- $ P(y = k) $ is the prior probability of class $ k $.

### Summary
- **Averaging Covariance Matrices**: Used when individual class covariance matrices are singular or unstable.
- **LDA vs. QDA**: Averaging leads to LDA, where all classes share the same covariance matrix, resulting in linear decision boundaries.
- **Discriminant Function**: Uses the averaged covariance matrix, making the means the primary class-specific parameters.

# Two-Class Example: Linear Discriminant Analysis (LDA)

## Introduction
In this example, we will explore Linear Discriminant Analysis (LDA) for a two-class classification problem. LDA is a classification algorithm that assumes that the data from each class is drawn from a Gaussian distribution with the same covariance matrix. This results in linear decision boundaries between classes.

## Key Concepts

### Mean Vectors
For each class $ k $, the mean vector $ \mu_k $ is calculated as:
$ \mu_k = \frac{1}{N_k} \sum_{i \in C_k} \mathbf{x}_i $
where $ N_k $ is the number of samples in class $ k $ and $ \mathbf{x}_i $ are the samples in class $ k $.

### Covariance Matrices
For each class $ k $, the covariance matrix $ \Sigma_k $ is calculated as:
$ \Sigma_k = \frac{1}{N_k - 1} \sum_{i \in C_k} (\mathbf{x}_i - \mu_k)(\mathbf{x}_i - \mu_k)^T $

### Averaged Covariance Matrix
The averaged covariance matrix $ \Sigma $ is calculated as:
$ \Sigma = \frac{1}{2} (\Sigma_1 + \Sigma_2) $
where $ \Sigma_1 $ and $ \Sigma_2 $ are the covariance matrices for class 1 and class 2, respectively.

### Discriminant Function
The discriminant function for class $ k $ is given by:
$ \delta_k(\mathbf{x}) = -\frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma^{-1} (\mathbf{x} - \mu_k) + \log P(y = k) $
where:
- $ \mu_k $ is the mean vector for class $ k $.
- $ \Sigma $ is the averaged covariance matrix.
- $ P(y = k) $ is the prior probability of class $ k $.

### Linear Discriminant Function
The linear discriminant function $ f(\mathbf{x}) $ is defined as:
$ f(\mathbf{x}) = \log P(y=1|\mathbf{x}) - \log P(y=2|\mathbf{x}) $
This can be expressed as:
$ f(\mathbf{x}) = \mathbf{w}^T \mathbf{x} + w_0 $
where:
- $ \mathbf{w} = \Sigma^{-1} (\mu_2 - \mu_1) $
- $ w_0 = -\frac{1}{2} \mu_2^T \Sigma^{-1} \mu_2 + \frac{1}{2} \mu_1^T \Sigma^{-1} \mu_1 $

## Summary
- **Mean Vectors**: Calculated for each class.
- **Covariance Matrices**: Calculated for each class.
- **Averaged Covariance Matrix**: Used to stabilize the covariance estimate.
- **Discriminant Function**: Uses the averaged covariance matrix, resulting in linear decision boundaries.
- **Linear Discriminant Function**: Expressed as a linear combination of the input features.

```python
# Comprehensive Example of LDA and QDA with Visualizations

import numpy as np
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate synthetic data
np.random.seed(0)
X = np.random.randn(200, 2)
y = np.array([0 if x[0]**2 + x[1]**2 < 1 else 1 for x in X])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Initialize the LDA and QDA classifiers
lda = LinearDiscriminantAnalysis()
qda = QuadraticDiscriminantAnalysis()

# Fit the models
lda.fit(X_train, y_train)
qda.fit(X_train, y_train)

# Predict the labels for the test set
y_pred_lda = lda.predict(X_test)
y_pred_qda = qda.predict(X_test)

# Calculate the accuracy
accuracy_lda = accuracy_score(y_test, y_pred_lda)
accuracy_qda = accuracy_score(y_test, y_pred_qda)

print(f"LDA Accuracy: {accuracy_lda}")
print(f"QDA Accuracy: {accuracy_qda}")

# Plot decision boundaries
def plot_decision_boundary(model, X, y, title):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plot_decision_boundary(lda, X, y, 'LDA Decision Boundary')

plt.subplot(1, 2, 2)
plot_decision_boundary(qda, X, y, 'QDA Decision Boundary')

plt.tight_layout()
plt.show()
```

## Simplified Covariance Matrix in Gaussian-Based Classification

In some cases, even a full averaged covariance matrix is too complex to estimate accurately, especially when the dataset is small or the number of features is large. A common simplification is to assume that all features have the same variance and are uncorrelated. This results in a diagonal covariance matrix where the off-diagonal elements are zero.

### Simplified Covariance Matrix

The simplified covariance matrix can be estimated by multiplying the standard deviation by the identity matrix. This approach assumes that the variance of each feature is the same and that there are no correlations between features.

### Comparison to Averaged Covariance Matrix

- **Averaged Covariance Matrix**: This matrix is calculated by averaging the covariance matrices of all classes. It captures the variances and covariances of the features, allowing for linear decision boundaries.
- **Simplified Covariance Matrix**: This matrix assumes equal variance for all features and no correlation between them. It simplifies the estimation process and results in a diagonal covariance matrix.

### Derivation of the Discriminant Function

Given the simplified covariance matrix, the discriminant function can be derived as follows:

1. **Covariance Matrix**:
$
\Sigma = \sigma^2 I
$
where $ \sigma^2 $ is the common variance and $ I $ is the identity matrix.

2. **Inverse of Covariance Matrix**:
$
\Sigma^{-1} = \frac{1}{\sigma^2} I
$

3. **Determinant of Covariance Matrix**:
$
 |\Sigma| = (\sigma^2)^d
$
where $ d $ is the number of features.

4. **Discriminant Function**:
$
\delta_k(\mathbf{x}) = -\frac{1}{2} \log |\Sigma| - \frac{1}{2} (\mathbf{x} - \mu_k)^T \Sigma^{-1} (\mathbf{x} - \mu_k) + \log P(y = k)
$

Substituting the simplified covariance matrix:
$
\delta_k(\mathbf{x}) = -\frac{1}{2} \log (\sigma^2)^d - \frac{1}{2} (\mathbf{x} - \mu_k)^T \left( \frac{1}{\sigma^2} I \right) (\mathbf{x} - \mu_k) + \log P(y = k)
$

Simplifying further:
$
\delta_k(\mathbf{x}) = -\frac{d}{2} \log \sigma^2 - \frac{1}{2\sigma^2} (\mathbf{x} - \mu_k)^T (\mathbf{x} - \mu_k) + \log P(y = k)
$

Since $ \sigma^2 $ is common for all classes, it can be ignored when comparing discriminant functions:
    $
\delta_k(\mathbf{x}) = - \frac{1}{2\sigma^2} (\mathbf{x} - \mu_k)^T (\mathbf{x} - \mu_k) + \log P(y = k)
$
